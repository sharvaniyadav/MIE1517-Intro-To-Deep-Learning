Download the anonymized data provided on Quercus. Split the data into training, validation, and test sets.

```python
# Part B.1 — Data Loading and Subject-Based Splitting
# In this section, the dataset was unzipped, loaded, split by subject into train/val/test (70/15/15),
# the class counts were shown, the split was saved, DataLoaders were created, and a quick check was done for overlap.

# Step 1: Importing Necessary Libraries
# These libraries were used for file handling, data loading, transforms, and splitting by subject.
import os, json, random, collections, zipfile
from pathlib import Path

import numpy as np
import torch
from torch.utils.data import Subset, DataLoader
from torchvision import datasets, transforms
from sklearn.model_selection import GroupShuffleSplit

# Step 2: Unzipping the Dataset and Setting the Root Path
# The dataset zip file was already in the Colab environment.
# It was unzipped into /content if it wasn’t already, and the root path was set.
zip_path = "/content/A2_Hand_Gesture_Dataset_revised.zip"
unzip_root = "/content"

if not os.path.exists(os.path.join(unzip_root, "Lab_3b_Gesture_Dataset Revised")):
    with zipfile.ZipFile(zip_path, 'r') as zf:
        zf.extractall(unzip_root)
    print("[INFO] Dataset unzipped.")
else:
    print("[INFO] Dataset already unzipped.")

DATA_ROOT = "/content/Lab_3b_Gesture_Dataset Revised"
print(f"[INFO] DATA_ROOT = {DATA_ROOT}")

# Step 3: Setting Random Seeds and Defining Transforms
# Fixed seeds were set to make sure the split is always the same.
# A simple transform was used to resize images and convert them to tensors.
seed = 42
random.seed(seed)
np.random.seed(seed)
torch.manual_seed(seed)

transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
])

# Step 4: Loading the Dataset
# The dataset had folders for each class (A–I).
# ImageFolder assigned labels based on these folders.
full_ds = datasets.ImageFolder(root=DATA_ROOT, transform=transform)
class_names = full_ds.classes
print("[INFO] Classes:", class_names)
print("[INFO] Total Images:", len(full_ds))

# Step 5: Extracting Subject IDs
# Each filename had a subject ID at the start, before the first underscore.
# These IDs were used to split the data by subject.
def get_subject_id(path_str):
    stem = Path(path_str).stem
    return stem.split("_")[0].lower()

subjects = [get_subject_id(fp) for fp, _ in full_ds.samples]
unique_subjects = sorted(set(subjects))
print(f"[INFO] Number of Unique Subjects: {len(unique_subjects)}")

# Step 6: Splitting the Dataset by Subject (70/15/15)
# GroupShuffleSplit was used so that all images from one subject stay in the same set.
# 15% of subjects were used for test, and the rest were split into train and val.
idx_all = np.arange(len(full_ds))
groups = np.array(subjects)

# Test split
gss1 = GroupShuffleSplit(n_splits=1, test_size=0.15, random_state=seed)
trainval_idx, test_idx = next(gss1.split(idx_all, groups=groups))

# Validation split
val_rel = 0.15 / (1 - 0.15)
gss2 = GroupShuffleSplit(n_splits=1, test_size=val_rel, random_state=seed)
train_idx_rel, val_idx_rel = next(gss2.split(trainval_idx, groups=groups[trainval_idx]))

train_idx = trainval_idx[train_idx_rel]
val_idx = trainval_idx[val_idx_rel]

print(f"[INFO] Split Sizes -> Train: {len(train_idx)} | Val: {len(val_idx)} | Test: {len(test_idx)}")

# Step 7: Showing Class Counts and Saving the Split
# The number of images per class was printed for each split.
# The split indices were saved to a JSON file to reuse later.
def counts_by_class(ds, idxs):
    counter = collections.Counter(int(ds[i][1]) for i in idxs)
    return {ds.classes[k]: counter.get(k, 0) for k in range(len(ds.classes))}

def summarize(name, idxs):
    print(f"\n[{name.upper()}] Total Images: {len(idxs)}")
    counts = counts_by_class(full_ds, idxs)
    for c in full_ds.classes:
        print(f"  {c}: {counts[c]}")
    return counts

train_counts = summarize("train", train_idx)
val_counts = summarize("val", val_idx)
test_counts = summarize("test", test_idx)

split_file = "/content/subject_disjoint_split.json"
with open(split_file, "w") as f:
    json.dump({
        "classes": class_names,
        "train_idx": train_idx.tolist(),
        "val_idx": val_idx.tolist(),
        "test_idx": test_idx.tolist(),
        "seed": seed
    }, f, indent=2)
print(f"\n[INFO] Split saved to {split_file}")

# Step 8: Creating Subsets and DataLoaders
# Subsets were made for each split, and DataLoaders were created to load data during training.
train_ds = Subset(full_ds, train_idx)
val_ds = Subset(full_ds, val_idx)
test_ds = Subset(full_ds, test_idx)

def make_loader(ds, bs=32, shuffle=False, nw=2):
    return DataLoader(ds, batch_size=bs, shuffle=shuffle, num_workers=nw, pin_memory=True)

batch_size = 32
train_loader = make_loader(train_ds, bs=batch_size, shuffle=True)
val_loader = make_loader(val_ds, bs=batch_size)
test_loader = make_loader(test_ds, bs=batch_size)
print("[INFO] DataLoaders Ready.")

# Step 9: Checking for Subject Overlap
# A quick check was done to make sure subjects don’t appear in more than one split.
def subject_set(subset):
    return set(get_subject_id(full_ds.samples[i][0]) for i in subset.indices)

subj_train = subject_set(train_ds)
subj_val = subject_set(val_ds)
subj_test = subject_set(test_ds)

print(f"[CHECK] Train ∩ Val = {len(subj_train & subj_val)}")
print(f"[CHECK] Train ∩ Test = {len(subj_train & subj_test)}")
print(f"[CHECK] Val ∩ Test = {len(subj_val & subj_test)}")
```

## Part 2. Model Building and Sanity Checking [5pt]

### Part (i) Convolutional Network [2pt MODEL]

Build a convolutional neural network model that takes the (224x224 RGB) image as input, and predicts the gesture
letter. Your model should be a subclass of nn.Module. Explain your choice of neural network architecture: how
many layers did you choose? What types of layers did you use? Were they fully-connected or convolutional?
What about other decisions like pooling layers, activation functions, number of channels / hidden units?

```python
# PART B.2(i) — BUILDING THE CONVOLUTIONAL NEURAL NETWORK MODEL
# This model took 224x224 RGB images as input and produced logits for 9 gesture classes (A–I)

import torch
import torch.nn as nn
import torch.nn.functional as F

class GestureCNN(nn.Module):
    def __init__(self, dropout_p: float = 0.4):
        super().__init__()
        # Block 1: Applied Conv → BN → ReLU → MaxPool (224x224x3 → 112x112x32)
        self.conv1 = nn.Conv2d(3, 32, kernel_size=3, padding=1)
        self.bn1   = nn.BatchNorm2d(32)

        # Block 2: Applied Conv → BN → ReLU → MaxPool (112x112x32 → 56x56x64)
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2   = nn.BatchNorm2d(64)

        # Block 3: Applied Conv → BN → ReLU → MaxPool (56x56x64 → 28x28x128)
        self.conv3 = nn.Conv2d(64, 128, kernel_size=3, padding=1)
        self.bn3   = nn.BatchNorm2d(128)

        # Used MaxPooling to reduce spatial size and Global Average Pooling to flatten to 128-D
        self.pool  = nn.MaxPool2d(2, 2)
        self.gap   = nn.AdaptiveAvgPool2d(1)

        # Defined Fully Connected Layers for final classification (128 → 128 → 9)
        self.fc1   = nn.Linear(128, 128)
        self.drop  = nn.Dropout(dropout_p)
        self.fc2   = nn.Linear(128, 9)

    def forward(self, x):
        # Passed input through each block
        x = self.pool(F.relu(self.bn1(self.conv1(x))))  # Block 1
        x = self.pool(F.relu(self.bn2(self.conv2(x))))  # Block 2
        x = self.pool(F.relu(self.bn3(self.conv3(x))))  # Block 3

        # Applied Global Average Pooling and flattened output
        x = self.gap(x).flatten(1)

        # Passed through FC layers with ReLU and Dropout
        x = F.relu(self.fc1(x))
        x = self.drop(x)
        x = self.fc2(x)
        return x

# Quick shape check to verify output dimension was correct
if __name__ == "__main__":
    model = GestureCNN()
    out = model(torch.randn(2, 3, 224, 224))
    print("Output shape:", out.shape)
```

```python
'''
PROVIDE YOUR ANSWER BELOW

The model used is a custom convolutional neural network built as a subclass of nn.Module. It takes 224×224
RGB images as input and predicts one of 9 gesture classes.
The architecture consists of three convolutional layers followed by two fully connected layers.
Each convolutional layer uses 3×3 filters, with the number of channels
increasing from 32 to 64 to 128 as the network goes deeper.
This allows the model to learn simple features in the early layers and more complex patterns in later layers.

After each convolutional layer, ReLU activation is applied to introduce non-linearity,
and a 2×2 max pooling layer is used to reduce the spatial dimensions by half.
This helps control the number of parameters and makes the network more efficient. After the final convolution block,
global average pooling is applied to reduce the 28×28×128 feature map into a 128-dimensional vector.
This avoids flattening a large feature map and keeps the fully connected layers smaller and easier to train.

The classifier consists of two fully connected layers with 128 hidden units and dropout for regularization.
The final layer outputs 9 logits, one for each gesture class (A–I).
The design is simple but effective: convolutional layers extract spatial features,
ReLU provides non-linearity, max pooling reduces size, and the fully connected layers perform the classification.

'''
```

### Part (ii) Training Code [2pt MODEL]

Write code that trains your neural network given some training data. Your training code should make it easy
to tweak the usual hyperparameters, like batch size, learning rate, and the model object itself. Make sure
that you are checkpointing your models from time to time (the frequency is up to you). Explain your choice
of loss function and optimizer.

```python
# PART B.2(ii) — TRAINING CODE (EASY HYPERPARAMETERS + CHECKPOINTS)
# This code trained the CNN on the subject-disjoint split and saved checkpoints during training.

import torch
import torch.nn as nn
from torch.optim import Adam
from torch.optim.lr_scheduler import StepLR
from pathlib import Path
import time

# Step 1: Defined loss function and a small accuracy helper
criterion = nn.CrossEntropyLoss()

def batch_accuracy(logits, targets):
    preds = logits.argmax(1)
    return (preds == targets).float().mean().item()

# Step 2: Wrote a single-epoch training loop
def train_one_epoch(model, loader, optimizer, device):
    model.train()
    total_loss, total_acc, total_n = 0.0, 0.0, 0
    for x, y in loader:
        x, y = x.to(device), y.to(device)

        optimizer.zero_grad()
        logits = model(x)
        loss = criterion(logits, y)
        loss.backward()
        optimizer.step()

        bs = y.size(0)
        total_loss += loss.item() * bs
        total_acc  += batch_accuracy(logits, y) * bs
        total_n    += bs
    return total_loss / total_n, total_acc / total_n

# Step 3: Wrote an evaluation loop (no gradients)
@torch.no_grad()
def evaluate(model, loader, device):
    model.eval()
    total_loss, total_acc, total_n = 0.0, 0.0, 0
    for x, y in loader:
        x, y = x.to(device), y.to(device)
        logits = model(x)
        loss = criterion(logits, y)

        bs = y.size(0)
        total_loss += loss.item() * bs
        total_acc  += batch_accuracy(logits, y) * bs
        total_n    += bs
    return total_loss / total_n, total_acc / total_n

# Step 4: Wrapped training with hyperparameters and checkpoint saving
def train_model(
    model,
    train_loader,
    val_loader,
    device,
    *,
    epochs=15,            # easy to change
    lr=3e-4,              # easy to change
    weight_decay=1e-4,    # L2 regularization
    step_size=5,          # drop LR every 'step_size' epochs
    gamma=0.5,            # multiply LR by this factor
    ckpt_dir="/content/checkpoints",
    ckpt_every=5          # save every N epochs
):
    optimizer = Adam(model.parameters(), lr=lr, weight_decay=weight_decay)
    scheduler = StepLR(optimizer, step_size=step_size, gamma=gamma)

    Path(ckpt_dir).mkdir(parents=True, exist_ok=True)
    best_val_acc, best_path = 0.0, None

    for epoch in range(1, epochs + 1):
        t0 = time.time()
        train_loss, train_acc = train_one_epoch(model, train_loader, optimizer, device)
        val_loss,   val_acc   = evaluate(model, val_loader, device)
        scheduler.step()
        dt = time.time() - t0

        print(f"Epoch {epoch:02d} | "
              f"Train Loss {train_loss:.4f} Acc {train_acc:.2%} | "
              f"Val Loss {val_loss:.4f} Acc {val_acc:.2%} | "
              f"{dt:.1f}s")

        # Saved periodic checkpoint
        if epoch % ckpt_every == 0:
            path = f"{ckpt_dir}/gesturecnn_epoch{epoch:02d}.pt"
            torch.save({
                "epoch": epoch,
                "model_state": model.state_dict(),
                "optimizer_state": optimizer.state_dict(),
                "scheduler_state": scheduler.state_dict()
            }, path)
            print(f"[INFO] Saved checkpoint to {path}")

        # Saved best model based on validation accuracy
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_path = f"{ckpt_dir}/gesturecnn_best.pt"
            torch.save({
                "epoch": epoch,
                "model_state": model.state_dict(),
                "optimizer_state": optimizer.state_dict(),
                "scheduler_state": scheduler.state_dict()
            }, best_path)
            print(f"[INFO] Updated best model (Val Acc {best_val_acc:.2%})")

    print(f"[INFO] Best validation accuracy: {best_val_acc:.2%}")
    return best_path

# Step 5: Chose device, created model, and started training (hyperparameters are easy to tweak)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = GestureCNN().to(device)

best_ckpt = train_model(
    model,
    train_loader,
    val_loader,
    device,
    epochs=15,
    lr=3e-4,
    weight_decay=1e-4,
    step_size=5,
    gamma=0.5,
    ckpt_dir="/content/checkpoints",
    ckpt_every=5
)
print("Best checkpoint path:", best_ckpt)
```

### Part (iii) “Overfit” to a Small Dataset - [1pt MODEL]

One way to sanity check our neural network model and training code is to check whether the model is capable
of “overfitting” or “memorizing” a small dataset. A properly constructed CNN with correct training code
should be able to memorize the answers to a small number of images quickly.

Construct a small dataset (e.g. just the images that you have collected). Then show that your model and
training code is capable of memorizing the labels of this small data set.

With a large batch size (e.g. the entire small dataset) and learning rate that is not too high, you should be
able to obtain a 100% training accuracy on that small dataset relatively quickly (within 200 iterations).

```python
# PART B.2(iii) — "OVERFIT" TO A SMALL DATASET (SANITY CHECK)
# This section created a tiny training set and showed that the model could memorize it quickly.

import torch
from torch.utils.data import Subset, DataLoader
import numpy as np

# Step 1: Built a tiny dataset from the subject-disjoint training split
# Took the first 64 samples to keep it small and mixed across classes.
small_indices = np.arange(min(64, len(train_ds)))   # train_ds came from Part 1
small_ds = Subset(train_ds, small_indices)

# Step 2: Used a full-batch loader (the entire small dataset in one batch)
small_loader = DataLoader(
    small_ds,
    batch_size=len(small_ds),   # full batch
    shuffle=True,
    num_workers=0,
    pin_memory=True
)

# Step 3: Initialized a fresh model and optimizer for the overfit test
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model_overfit = GestureCNN().to(device)
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model_overfit.parameters(), lr=1e-3, weight_decay=0.0)  # no wd to make memorization easier

# Step 4: Trained up to 200 iterations and stopped early at 100% train accuracy
def train_overfit(model, loader, max_iters=200):
    model.train()
    it = 0
    for _ in range(max_iters):
        for x, y in loader:
            x, y = x.to(device), y.to(device)

            optimizer.zero_grad()
            logits = model(x)
            loss = criterion(logits, y)
            loss.backward()
            optimizer.step()

            with torch.no_grad():
                preds = logits.argmax(1)
                acc = (preds == y).float().mean().item()

            it += 1
            if it % 10 == 0 or acc == 1.0:
                print(f"Iter {it:03d} | Loss: {loss.item():.4f} | Train Acc: {acc:.2%}")
            if acc == 1.0:
                print(f"[INFO] Reached 100% training accuracy at iteration {it}.")
                return
    print("[INFO] Reached max iterations without hitting 100%.")

train_overfit(model_overfit, small_loader, max_iters=200)
```

For many image classification tasks, it is generally not a good idea to train a very large deep neural network
model from scratch due to the enormous compute requirements and lack of sufficient amounts of training
data.

One of the better options is to try using an existing model that performs a similar task to the one you need
to solve. This method of utilizing a pre-trained network for other similar tasks is broadly termed **Transfer
Learning**. In this assignment, we will use Transfer Learning to extract features from the hand gesture
images. Then, train a smaller network to use these features as input and classify the hand gestures.

As you have learned from the CNN lecture, convolution layers extract various features from the images which
get utilized by the fully connected layers for correct classification. AlexNet architecture played a pivotal
role in establishing Deep Neural Nets as a go-to tool for image classification problems and we will use an
ImageNet pre-trained AlexNet model to extract features in this assignment.

```python
# PART B.4(i) — ALEXNET FEATURE EXTRACTION (FINAL)

# STEP 1: IMPORTS
import os, torch
from torchvision import models, datasets, transforms
from torch.utils.data import DataLoader
from pathlib import Path

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# STEP 2: LOADED ALEXNET AND KEPT ONLY CONV FEATURES
alexnet = models.alexnet(weights=models.AlexNet_Weights.IMAGENET1K_V1).to(device)
alexnet.eval()
feat = alexnet.features

# STEP 3: USED IMAGENET NORMALIZATION (ALEXNET WAS TRAINED WITH THIS)
imgnet_tfms = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize([0.485, 0.456, 0.406],
                         [0.229, 0.224, 0.225]),
])

# STEP 4: CHECKED IF TRAIN/VAL/TEST LOADERS ALREADY EXISTED
def loaders_exist():
    return all(name in globals() for name in ["train_loader", "val_loader", "test_loader"])

# STEP 5: FOUND THE DATA FOLDER WHEN LOADERS WERE MISSING
def find_data_root():
    # TRIED COMMON NAMES FIRST
    candidates = [
        "/content/Processed_Lab_3b_Gesture_Dataset_Revised",
        "/content/Processed_Lab_3b_Gesture_Dataset Revised",
        "/content/Lab_3b_Gesture_Dataset_Revised",
        "/content/Lab_3b_Gesture_Dataset Revised",
        "/content/Gesture_Dataset",
        "/content/gesture_data",
    ]
    for root in candidates:
        if os.path.isdir(root):
            has_train = os.path.isdir(os.path.join(root, "train"))
            has_val   = os.path.isdir(os.path.join(root, "val")) or os.path.isdir(os.path.join(root, "valid"))
            has_test  = os.path.isdir(os.path.join(root, "test"))
            if has_train and has_val and has_test:
                return root
    # LAST RESORT: SCANNED /CONTENT
    for root, dirs, _ in os.walk("/content"):
        d = set(n.lower() for n in dirs)
        if "train" in d and ("val" in d or "valid" in d) and "test" in d:
            return root
    return None

# STEP 6: BUILT DATALOADERS IF NOT ALREADY AVAILABLE
if not loaders_exist():
    DATA_ROOT = find_data_root()
    if DATA_ROOT is None:
        print("[ERROR] DID NOT FIND train/val(or valid)/test FOLDER UNDER /content")
        os.system("ls -lah /content")
        raise SystemExit

    TRAIN_DIR = os.path.join(DATA_ROOT, "train")
    VAL_DIR   = os.path.join(DATA_ROOT, "val") if os.path.isdir(os.path.join(DATA_ROOT, "val")) else os.path.join(DATA_ROOT, "valid")
    TEST_DIR  = os.path.join(DATA_ROOT, "test")

    train_ds = datasets.ImageFolder(TRAIN_DIR, transform=imgnet_tfms)
    val_ds   = datasets.ImageFolder(VAL_DIR,   transform=imgnet_tfms)
    test_ds  = datasets.ImageFolder(TEST_DIR,  transform=imgnet_tfms)

    train_loader = DataLoader(train_ds, batch_size=64, shuffle=False, num_workers=2, pin_memory=True)
    val_loader   = DataLoader(val_ds,   batch_size=64, shuffle=False, num_workers=2, pin_memory=True)
    test_loader  = DataLoader(test_ds,  batch_size=64, shuffle=False, num_workers=2, pin_memory=True)

# STEP 7: RAN ONE BATCH THROUGH ALEXNET.FEATURES TO SHOW SHAPE
xb, _ = next(iter(train_loader))
with torch.no_grad():
    fb = feat(xb.to(device))                  # [N, 256, 6, 6]
print("Feature batch shape:", tuple(fb.shape))

# STEP 8: EXTRACTED AND SAVED FEATURES ONCE FOR EACH SPLIT
@torch.no_grad()
def extract_and_save(loader, out_path: Path):
    alexnet.eval()
    feats, labels = [], []
    for x, y in loader:
        f = feat(x.to(device)).cpu()          # [B, 256, 6, 6]
        feats.append(f)
        labels.append(y.clone())
    feats  = torch.cat(feats, dim=0)          # [TOTAL, 256, 6, 6]
    labels = torch.cat(labels, dim=0)         # [TOTAL]
    torch.save({"features": feats, "labels": labels}, out_path)

OUT_DIR = Path("/content/alexnet_features")
OUT_DIR.mkdir(parents=True, exist_ok=True)

extract_and_save(train_loader, OUT_DIR / "train_features.pt")
extract_and_save(val_loader,   OUT_DIR / "val_features.pt")
extract_and_save(test_loader,  OUT_DIR / "test_features.pt")

print("Saved:", OUT_DIR / "train_features.pt")
print("Saved:", OUT_DIR / "val_features.pt")
print("Saved:", OUT_DIR / "test_features.pt")
```

```python
# PART 5(ii) — Comparing Baseline CNN and Transfer Learning on My Collected Images
# also added a simple visualization to double check predictions

import os
import math
import torch
import pandas as pd
import matplotlib.pyplot as plt
from torchvision import models, transforms
import torch.nn as nn

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# step 1: mapped the letters A–I to numeric labels and extracted true labels from the filenames
label_map = {'A':0, 'B':1, 'C':2, 'D':3, 'E':4, 'F':5, 'G':6, 'H':7, 'I':8}
idx_to_letter = {v: k for k, v in label_map.items()}

true_labels = []
for p in image_paths:
    name = os.path.basename(p)
    letter = name.split("_")[1].upper()  # pulled the letter part
    true_labels.append(label_map[letter])

true_labels = torch.tensor(true_labels, dtype=torch.long, device=device)

# step 2: sent the collected image tensor to the device
collected_tensor = collected_tensor.to(device)

# step 3: loaded the baseline CNN model (Part 3) and got predictions on my images
class GestureCNN(nn.Module):
    def __init__(self, dropout_p=0.3):
        super().__init__()
        self.net = nn.Sequential(
            nn.Conv2d(3, 32, 3, padding=1),
            nn.BatchNorm2d(32),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1),
            nn.BatchNorm2d(64),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.BatchNorm2d(128),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Flatten(),
            nn.Linear(128 * 28 * 28, 256),
            nn.ReLU(),
            nn.Dropout(dropout_p),
            nn.Linear(256, 9)
        )

    def forward(self, x):
        return self.net(x)

baseline_model = GestureCNN(dropout_p=0.3).to(device)
baseline_ckpt = torch.load("/content/checkpoints/gesturecnn_best.pt", map_location=device)
baseline_model.load_state_dict(baseline_ckpt["model_state"], strict=False)
baseline_model.eval()

@torch.no_grad()
def get_preds_and_acc(model, imgs, labels):
    logits = model(imgs)
    preds = logits.argmax(1)
    acc = (preds == labels).float().mean().item() * 100.0
    return preds, acc

baseline_preds, baseline_acc = get_preds_and_acc(baseline_model, collected_tensor, true_labels)

# step 4: normalized the collected images, extracted AlexNet features, and evaluated the transfer head
imgnet_norm = transforms.Normalize([0.485, 0.456, 0.406],
                                   [0.229, 0.224, 0.225])

collected_norm = collected_tensor.clone()
for i in range(collected_norm.size(0)):
    collected_norm[i] = imgnet_norm(collected_norm[i])

alex = models.alexnet(weights=models.AlexNet_Weights.IMAGENET1K_V1).to(device)
alex.eval()
features_block = alex.features

@torch.no_grad()
def extract_feats(x):
    return features_block(x)

feats = extract_feats(collected_norm)

class GestureClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(256 * 6 * 6, 512)
        self.relu = nn.ReLU()
        self.drop = nn.Dropout(0.2)
        self.fc2 = nn.Linear(512, 9)
    def forward(self, x):
        x = self.flatten(x)
        x = self.fc1(x)
        x = self.relu(x)
        x = self.drop(x)
        x = self.fc2(x)
        return x

transfer_head = GestureClassifier().to(device)
transfer_ckpt = torch.load("/content/alexnet_features/gesture_classifier.pt", map_location=device)
if isinstance(transfer_ckpt, dict) and "model_state" in transfer_ckpt:
    transfer_head.load_state_dict(transfer_ckpt["model_state"])
else:
    transfer_head.load_state_dict(transfer_ckpt)
transfer_head.eval()

transfer_preds, transfer_acc = get_preds_and_acc(transfer_head, feats, true_labels)

# step 5: summarized the original test accuracy vs accuracy on my collected images in a DataFrame
baseline_test_acc = 40.43  # from Part 3(iv)
transfer_test_acc = 88.68  # from Part 4(iv)

df = pd.DataFrame({
    "Model": ["Baseline CNN (Part 3)", "Transfer Learning (Part 4)"],
    "Test Accuracy (Original)": [baseline_test_acc, transfer_test_acc],
    "Accuracy on Collected Images": [baseline_acc, transfer_acc],
    "Difference": [baseline_acc - baseline_test_acc, transfer_acc - transfer_test_acc],
})

print("\nSummary (original test acc vs my collected images):")
print(df.to_string(index=False))

# step 6: visualized all 27 collected images with true, baseline, and transfer predictions
num_images = len(image_paths)
cols = 6
rows = math.ceil(num_images / cols)

plt.figure(figsize=(18, 3 * rows))

for i in range(num_images):
    img = collected_tensor[i].cpu().permute(1, 2, 0)
    true_letter = idx_to_letter[int(true_labels[i].cpu())]
    base_pred_letter = idx_to_letter[int(baseline_preds[i].cpu())]
    trans_pred_letter = idx_to_letter[int(transfer_preds[i].cpu())]

    plt.subplot(rows, cols, i+1)
    plt.imshow(img)
    plt.axis('off')
    plt.title(f"T:{true_letter} | B:{base_pred_letter} | X:{trans_pred_letter}", fontsize=9)

plt.suptitle("T=true, B=Baseline, X=Transfer — Predictions on All Collected Images", fontsize=14)
plt.tight_layout()
plt.show()
```
